```python
import pandas as pd

df = pd.read_json('../data/processed/streaming_users_clean.json')
print("shape", df.shape)
```

## Resumen del proceso ##

Este proyecto analizó un dataset de usuarios de una plataforma de streaming que contenía 8.160 registros y 8 variables originales. A lo largo de cinco etapas.
Con el siguiente orden:

### Inspeccion inicial ###

    ► Inspeccion inicial
         → Duplicados en "user_id"
         → Encontramos valores nulos en la columnas: "last_login_date", "favorite_genre" y en "monthly_watch_time_mins"     
         → Valores imposibles en columnas "age", "monthly_watch_time_mins", "customer_support_tickets" y "last_login_date"
         → Ademas de edades imposibles o bastante improbables en "age" (-5 años, 150 años)
         → Ademas de valores no estandarizados en "country", "favorite_genre" y "suscription_plan"
    
    ► Calidad y limpieza
         → Eliminamos los duplicados de "user_id"
         → Eliminar los valores superior a la fecha "06/07/2026" de "last_login_date" y reemplazar los valores nulo por "(--/--/--)
         → Eliminar los valores superiores a 5000 minutos en "monthly_watch_time_mins" 
         → Estandarizar los valores de "favorite_genre" y reemplazar los nulos por "sin genero favorito" 
         → Filtrar las edades en un rango de 4-95 años, y eliminar el resto
         → Eliminar los valores imposibles de "customer_support_tickets" (solo los negativos)
         → Estandarizar los valores de "country" y "suscription_plan"

### Hallazgos principales (analisis exploratorio) ### 

Distribución de Edades

La distribución de edades de los usuarios revela que la plataforma atrae principalmente a un público adulto joven, con una edad promedio de 34.5 años y una mediana de 33 años, lo que indica que la mayoría de los usuarios se concentran entre los 20 y 40 años. Este patrón es consistente con lo que suele observarse en plataformas de streaming, donde el contenido digital tiene mayor penetración entre generaciones más jóvenes y millennials. El rango etario va desde los 4 hasta los 95 años, aunque los extremos son poco frecuentes, lo que confirma que el núcleo de la audiencia se encuentra en la adultez temprana y media.

Planes de Suscripción

En cuanto a los planes de suscripción, el plan Básico es el más utilizado, concentrando aproximadamente el 46% de los usuarios, seguido por el plan Estándar con un 31% y el plan Premium con un 23%. Esta predominancia del plan más económico sugiere que los usuarios son sensibles al precio y que la plataforma podría beneficiarse de estrategias para incentivar la migración hacia planes superiores. La diferencia entre el plan Básico y el Premium es considerable, lo que podría indicar que los usuarios no perciben un valor diferencial suficiente en los beneficios adicionales de los planes más caros, o que el costo es una barrera significativa.

Distribución por País

El análisis por país muestra que Argentina y Brasil son los mercados más importantes para la plataforma, concentrando la mayor cantidad de suscripciones con aproximadamente 2.500 y 2.400 usuarios respectivamente. México le sigue con cerca de 2.100 usuarios, mientras que Perú, Colombia, Chile y Uruguay completan el listado con cifras decrecientes. Esta concentración en los países más poblados de la región es esperable, pero también sugiere que la plataforma tiene una presencia más fuerte en el Cono Sur que en otras áreas de Latinoamérica. Argentina y Brasil deberían ser considerados mercados prioritarios para cualquier estrategia de crecimiento o retención.

Relación entre Edad y Minutos Visualizados

La relación entre la edad y el tiempo de visualización mensual revela una tendencia clara: los usuarios de mayor edad tienden a consumir más contenido. El coeficiente de correlación positivo, aunque moderado, indica que a medida que aumenta la edad, también lo hace la cantidad de minutos visualizados al mes. El pico de consumo se observa en el rango de 45 a 54 años, donde los usuarios superan los 1000 minutos mensuales en promedio. Los usuarios más jóvenes, especialmente aquellos menores de 20 años, son los que menos contenido consumen. Este hallazgo sugiere que los adultos mayores tienen más tiempo disponible o prefieren el streaming como principal fuente de entretenimiento, mientras que los jóvenes podrían estar distribuyendo su tiempo entre múltiples plataformas o actividades.

Preferencias de Género por País

Las preferencias de género varían significativamente según el país, reflejando diferencias culturales en los gustos audiovisuales. En Argentina, el Drama y el Thriller son los géneros más populares, lo que indica una preferencia por tramas intensas y narrativas complejas. En Brasil, el Romance y la Comedia tienen una fuerte presencia, lo que sugiere un gusto por historias sentimentales y contenido ligero. México se inclina hacia la Comedia y la Acción, mostrando una preferencia por el entretenimiento dinámico y humorístico. Chile y Perú comparten patrones similares, con Drama y Comedia como géneros predominantes. Colombia presenta una distribución más equilibrada entre géneros, sin una preferencia marcada por uno en particular. Finalmente, Uruguay muestra una particular inclinación hacia los Documentales y la Comedia, lo que podría indicar un público más exigente o con intereses culturales específicos. Estas diferencias subrayan la importancia de adaptar el catálogo de contenido a las preferencias locales para maximizar el engagement y la satisfacción del usuario en cada mercado.



### Interpretacion del pca ###

#### Varianza ####

El análisis de componentes principales revela que las tres componentes explican proporciones muy similares de la varianza total: PC1 explica el 33.68%, PC2 el 33.42% y PC3 el 32.90%. Esta distribución casi uniforme indica que las variables originales (edad, tiempo de visualización y tickets de soporte) no están correlacionadas entre sí, por lo que PCA no logra reducir la dimensionalidad sin pérdida significativa de información. Retener solo 2 componentes implicaría conservar apenas el 67.1% de la información original, lo cual no representa una reducción útil en términos prácticos. Para capturar el 100% de la varianza se requieren las 3 componentes originales. En este caso, el valor del PCA no reside en la compresión de los datos, sino en la comprensión de la estructura subyacente que cada componente revela sobre el comportamiento de los usuarios.

#### Loadings ####

La primera componente (PC1) está dominada por dos variables con pesos opuestos: el tiempo de visualización mensual contribuye positivamente con un loading de 0.74, mientras que los tickets de soporte lo hacen negativamente con -0.62. La edad tiene una contribución baja de 0.26. Esta configuración permite interpretar PC1 como una dimensión de "calidad de experiencia", donde los valores altos corresponden a usuarios que ven mucho contenido y reportan pocos problemas, mientras que los valores bajos reflejan usuarios con baja visualización y alta cantidad de tickets. Esta componente sugiere que existe una relación inversa entre el consumo y los problemas técnicos, lo que podría indicar que los usuarios satisfechos son aquellos que logran disfrutar del servicio sin interrupciones.

La segunda componente (PC2) está fuertemente dominada por la edad, con un loading de 0.86, acompañada por los tickets de soporte con 0.50 y el tiempo de visualización con apenas 0.12. PC2 puede interpretarse como una dimensión de "perfil etario", donde los valores altos corresponden a usuarios mayores que tienden a reportar más tickets, mientras que los valores bajos reflejan usuarios jóvenes con menos problemas. Esta componente revela que la edad es el factor más determinante para diferenciar a los usuarios, y que los adultos mayores podrían enfrentar más dificultades técnicas o tener menos tolerancia a los problemas.

### Limitaciones ###

Alcance geográfico limitado: El análisis se centra en 7 países latinoamericanos, por lo que los resultados pueden no ser generalizables a otras regiones.

Datos temporales: Se trabajó con datos de un solo momento (corte transversal), sin información de evolución temporal del comportamiento de los usuarios.

Datos faltantes: Algunos registros fueron imputados o eliminados, lo que podría introducir sesgos en el análisis.

Variables limitadas: El análisis se basa en 8 variables. Incorporar más variables (como tipo de dispositivo, horario de visualización, etc.) podría enriquecer el análisis.
